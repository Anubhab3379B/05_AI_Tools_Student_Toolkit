```python
# CELL 3: Student Workflow Automation
class StudentWorkflow:
    """Automate common student tasks with AI tool chains."""
    
    @staticmethod
    def research_pipeline(topic):
        """Automated research workflow."""
        steps = [
            f"1. Search Perplexity AI for '{topic}' with academic filter",
            f"2. Feed top 5 papers to Elicit for key findings extraction",
            f"3. Upload papers to NotebookLM for cross-referenced Q&A",
            f"4. Use Claude to draft literature review outline",
            f"5. Generate Anki flashcards for key concepts via ChatGPT",
        ]
        return steps
    
    @staticmethod
    def presentation_pipeline(notes_path):
        """Create a presentation from notes."""
        steps = [
            f"1. Upload {notes_path} to Notion AI for summarization",
            "2. Feed summary to Gamma.app for slide generation",
            "3. Enhance visuals in Canva with AI design suggestions",
            "4. Generate voiceover script with Claude",
            "5. Record narration with ElevenLabs voice synthesis",
        ]
        return steps

wf = StudentWorkflow()
print("Research Pipeline for 'Transformer Architecture':")
for step in wf.research_pipeline('Transformer Architecture'):
    print(f"  {step}")

print("\nPresentation Pipeline:")
for step in wf.presentation_pipeline('lecture_notes.pdf'):
    print(f"  {step}")
print("\nP5 AI Student Toolkit - COMPLETE!")
```
